夏目漱石の小説『吾輩は猫である』の文章（neko.txt）をMeCabを使って形態素解析し，その結果をneko.txt.mecabというファイルに保存せよ．

```python
# MeCabと辞書をインストール
# pip install mecab-python3

import MeCab

# テキストファイルをMeCabに通す
# https://www.codexa.net/mecab-python/

path = 'neko.txt'
text = "すもももももももものうち"
mecab = MeCab.Tagger('')
print(mecab.parse(text))
```

```python
path = './neko.txt'
with open(path, 'r', encoding='utf-8') as f:
    text = f.read().split()
parsed_text = []
for i in range(len(text)):
    parsed_text.append(mecab.parse(text[i]))
```

```python
save_path = 'neko.txt2.mecab'
with open(save_path, 'w', encoding='utf-8') as f:
    for text in parsed_text:
        f.write(text)
```

```python
# 採用した手法の方
import MeCab

# 表層形\t品詞,品詞細分類1,品詞細分類2,品詞細分類3,活用型,活用形,原形,読み,発音

mecab = MeCab.Tagger('')
file_path = 'neko.txt'
save_path = 'neko.txt.mecab'

with open(file_path, 'r', encoding='utf-8') as f1, open(save_path, 'w', encoding='utf-8') as f2:
    line = f1.readlines()
    for text in line:
        ans = mecab.parse(text)
        f2.write(ans)
```

30. 形態素解析結果の読み込み

形態素解析結果（neko.txt.mecab）を読み込むプログラムを実装せよ．ただし，各形態素は表層形（surface），基本形（base），品詞（pos），品詞細分類1（pos1）をキーとするマッピング型に格納し，1文を形態素（マッピング型）のリストとして表現せよ．第4章の残りの問題では，ここで作ったプログラムを活用せよ．

```python
import MeCab

# 表層形\t品詞,品詞細分類1,品詞細分類2,品詞細分類3,活用型,活用形,原形,読み,発音

mecab = MeCab.Tagger('')
file_path = 'neko.txt'
save_path = 'neko.txt.mecab'
texts = []
with open(file_path, 'r', encoding='utf-8') as f1, open(save_path, 'w', encoding='utf-8') as f2:
    line = f1.readlines()
    for text in line:
        ans = mecab.parse(text)
        f2.write(ans)
```
